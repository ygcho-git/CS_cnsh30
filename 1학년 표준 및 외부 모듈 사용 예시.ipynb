# $\sum$

$\displaystyle\sum_{i=1}^{100}i$ = $1 + 2 + 3 +... +  100$

````python
sum_v = 0
for i in range(1, 101) :
    sum_v = sum_v + i

print(sum_v)
````

$\displaystyle\sum_{i=1}^{n}x_i$ = $x_1 + x_2 + x_3 ... x_n$


````python
X = ["쓰레기", 2, 7, 3, 6, 8, 4]

sum_v = 0
for i in range(1, len(X)) :
    sum_v = sum_v + X[i]

print(sum_v)
````

$\displaystyle\sum_{i=0}^{n-1}x_i$ = $x_0 + x_1 + x_2 ... x_{n-1}$

````python
X = [2, 7, 3, 6, 8, 4]

sum_v = 0
for i in range(0, len(X)) :
    sum_v = sum_v + X[i]

print(sum_v)
````

$\displaystyle\sum_{i=0}^{n-1}a_ib_i$ = $a_0b_0 + a_1b_1 + a_2b_2 ... a_{n-1}b_{n-1}$

````python
a = [1, 3, 8, 5, 2]
b = [2, 7, 8, 3, 6]

sum_v = 0
for i in range(0, len(a)) :
    sum_v = sum_v + a[i] * b[i]

print(sum_v)
````

# numpy, matplotlib를 이용한 포물선 운동 궤적

````python
import numpy as np
import matplotlib.pyplot as plt

seta = int(input("입사각 = "))/180 * np.pi
v0 = 19.6 #float(input("처음 속력(m/s) = "))

g = 9.8
t = np.linspace(0, 100, 1000)
  # 0 부터 100까지 균일한 간격으로 1000개의 값 생성
  # [  0. 0.1001001 0.2002002 0.3003003 0.4004004 ... 99.8998999  100]

s = v0 * t * np.cos(seta)
h = v0 * t * np.sin(seta) - 1/2 * g * t * t

gak = seta * 180 / np.pi
ft = 2 * v0 * np.sin(seta) / g
max_h = v0 * v0 * np.sin(seta) * np.sin(seta) / (2 * g)
max_d = 2 * v0 * v0 * np.sin(seta) * np.cos(seta) / g

print("입사각(도) = ", round(gak, 0))
print("처음 속력(m/s) = ", v0)
print("지표면까지 걸린 시간(s) = ", round(ft, 3))
print("최고높이(m) = ", round(max_h, 3))
print("최대도달거리(m) = ", round(max_d, 3))

# 그래프 그리기
plt.plot(s, h)
plt.ylim(0, 20) # y축 범위
plt.xlim(0, 40) # x축 범위
plt.show()
````

# 파일 입출력
파일에 저장된 실험데이터(txt, csv, xls 등)을 읽어서, 파이썬으로 처리, 분석, 시각화(그래프)

````python
# 파일 쓰기

myfile = open("내 파일.txt", 'w')

for i in range(1, 11):
    data = f"{i}번째 줄입니다.\n"
    myfile.write(data)

myfile.close()
````

````python
# 파일 읽기

myfile = open("내 파일.txt", 'r')

for i in range(1, 11):
    data = myfile.readline()
    print(data)

myfile.close()
````

> 연립 방정식 : $\begin{cases} x+2y=3  & \\ 3x-y=2 \end{cases}$

````python
from sympy import Symbol, solve

y = Symbol('y')
equation1 = x + 2 * y - 3
equation2 = 3 * x - y + 2
solve((equation1, equation2), dict=True)
````

````python
from diffusers import StableDiffusionPipeline
import torch

model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

prompt = "a photo of an astronaut riding a horse on mars"
image = pipe(prompt).images[0]

image.save("astronaut_rides_horse.png")
````
